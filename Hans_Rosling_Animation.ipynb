```python
import pandas as pd
import plotly.express as px
```

```python
gapminder = px.data.gapminder()
```

```python
gapminder.head()
```

```python
gapminder.tail()
```

```python
px.scatter(data_frame= gapminder, x= 'gdpPercap',y='lifeExp')
```

```python
px.scatter(data_frame= gapminder, x= 'gdpPercap',y='lifeExp', size='pop') #size defines population
```

```python
px.scatter(data_frame= gapminder,
           x= 'gdpPercap',
           y='lifeExp',
           size = 'pop',
           hover_name= 'country',
           color = 'country')
```

```python
px.scatter(data_frame= gapminder,
           x= 'gdpPercap',
           y='lifeExp',
           size = 'pop',
           hover_name= 'country',
           color = 'country',
           animation_frame = 'year',
           animation_group ='country')
```

```python
px.scatter(data_frame= gapminder,
           x= 'gdpPercap',
           y='lifeExp',
           size = 'pop',
           hover_name= 'country',
           color = 'country',
           animation_frame = 'year',
           animation_group ='country',
           range_x= [0,50000],
           range_y= [20, 90])
```
